```python
import random

def generate_random_array(debug=False, N=21):
    """Renvoie un tableau contenant toutes les valeurs entières de 0 (inclus)
    à N (exclus) rangées dans un ordre aléatoire

    Args:
        debug (boolean): quand debug est vrai, la fonction renvoie toujours le
                         même tableau afin de simplifier le débogage de vos
                         algorithmes de tri
        N (int): la taille du tableau à renvoyer

    Returns:
        list[int]: un tableau d'entiers, de taille N, non ordonné
    """

    if debug:
        return [3, 9, 7, 1, 6, 2, 8, 4, 5, 0]

    array = list(range(0, N))
    random.shuffle(array)

    return array
```

```python
N=5000#, 8000
tab=generate_random_array(False,N)  
```

```python
liste_N=[10,100,1000,2000,5000, 8000]
compl_bubble=[54,54,500499,2000999,  12502499 ,32003999 ]
compl_insertion=[39,37, 244131,1015639, 6217524 ,15962407]
compl_selection=[55,55,500500,2001000,12502500 ,32004000]
compl_sort=[25,54,500499,26021, 73302 ,124880]
```

```python
plt.plot(liste_N,compl_bubble, '*-', color='red',markersize=10)
plt.plot(liste_N,compl_insertion, '+-', color='pink',markersize=10)
plt.plot(liste_N,compl_selection, '*-', color='blue',markersize=10)
plt.plot(liste_N,compl_sort, '+-', color='yellow',markersize=10)

y=[i**2 for i in liste_N]
plt.plot(liste_N,y,'+-',color='black')

y2=[i*math.log(i) for i in liste_N]
plt.plot(liste_N,y2,'*-',color='green')

plt.title('complexité')
plt.xlabel('taille de la liste')
#plt.ylabel('hit maximal (nbre de fois où on passe par une ligne de code dans une fonction)')
plt.ylabel('hit maximal')
plt.legend(["Tri bubble","Tri par insertion","Tri par selection", "Tri rapide", "O(n²)","O(n log(n))"]) 
plt.show()
```
